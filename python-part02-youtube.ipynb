```python
hight, lower = 0, 9999
start = 1  

while True:
    while True: 
        user_input = input(f"ครั้งที่ {start} [พิมพ์ 'exit' เพื่อจบ]: ").lower().strip()
        
        # 1. เช็คก่อนว่าอยากออกไหม
        if user_input == 'exit' or user_input == 'stop':
            break 
            
        # 2. เช็คว่าเป็นตัวเลขหรือไม่
        if user_input.isdigit(): 
            number = int(user_input)
            break 
        else:
            print("กรุณาป้อนตัวเลข หรือพิมพ์ 'exit' เท่านั้น")

    # เช็คเงื่อนไขการออกจากลูปใหญ่
    if user_input == 'exit' or user_input == 'stop' or (user_input.isdigit() and int(user_input) < 0):
        break
    
    # คำนวณค่า (ใช้ตัวแปร number ที่เราแปลงไว้แล้ว)
    if number > hight:
        hight = number
    if number < lower:
        lower = number
    
    start += 1 

print("\n" + "="*20)
print(f"สรุปผลการป้อนข้อมูล {start-1} ครั้ง")
print(f"🚀 สูงสุด: {hight}")
print(f"📉 ต่ำสุด: {lower}")
print("="*20)
```

```python
import random

myrandom = random.randrange(1, 7)  # สุ่มเลขระหว่าง 1 ถึง 6
n = 1  # กำหนดจำนวน loop ในแต่ละรอบ
print("คุณมีโอกาส 3 ครั้งในการตอบคำถาม")

while n <= 3:  # กำหนดให้มีการทายไม่เกิน 3 ครั้ง
    try:
        number = int(input(f"ป้อนจำนวนครั้งที่ {n}: "))  # รับค่าจากผู้เล่น
        correct = (number == myrandom)  # เช็คว่าถูกต้องไหม

        if correct:
            print("ขอแสดงความยินดีด้วย คุณทายถูกต้อง!")
            break  # ถ้าทายถูกให้จบเกม
        else:
            if number > myrandom:
                print("ค่าที่ถูกต้องน้อยกว่านี้")
            elif number < myrandom:
                print("ค่าที่ถูกต้องมากกว่านี้")
        n += 1  # เพิ่มจำนวนครั้งที่ทาย

    except ValueError:
        print("กรุณาป้อนตัวเลขที่ถูกต้อง")

else:  # ถ้าทายครบ 3 ครั้งแล้วยังไม่ถูกต้อง
    print("ขอแสดงความเสียใจด้วยนะ คุณพลาดแล้วละ")
    print(f"เฉลย: {myrandom}")
```

```python
import random
myrandom = random.randrange(1,7)
n=1 #กำหนดจำนวน loop แต่ละรอบให้
correct=False
print("คุณมีโอกาส 3 ครั้งในการตอบคำถาม")
while True:
    number = int(input(f"ป้อนจำนวนครั้งที่ {n}: "))
    correct=(number==myrandom) #ผลลัพธ์จะเป็น true / false
    if not correct:
        if number > myrandom:
            print("ค่าที่ถูกต้องน้อยกว่านี้")
        if number < myrandom:
            print("ค่าที่ถูกต้องมากกว่านี้")
    if correct:
        print("ขอแสดงความยินดีด้วย คุณมีความสามารถ")
        break
    if number < 0 or n==3:
        break
    n+=1
if not correct:
    print("ขอแสดงความเสียใจด้วยนะ คุณพลาดแล้วละ")
    print(f"เฉลย:{myrandom}")
```
