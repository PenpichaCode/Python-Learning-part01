```python
# if ซ้อน if
# Create Bank service
Name = input("ชื่อ:")
password = input("กรอกรหัสผ่าน:")

if Name == "member" and password == "1234":
    print("เข้าสู่ระบบสำเร็จ")
    service = input("ป้อนหมายเลขบริการ(1-2):")
    if service == "1": #if ย่อย จำทำงานก็ต่อเมื่อ if หลักเป็นจริง
        print("ถอนเงิน")
    elif service == "2":
        print("ฝากเงิน")
    else:
        print("หมายเลขบริการไม่ถูกต้อง")
else:
    print("ไม่พบบัญชี")
```

```python
#คล้าย if แต่สามารถกำหนดทางเลือกได้มากกว่า
# case function
#บริการธนาคาร
#syntax 
# case ค่าที 1:
#     satatment
# case ค่าที N:
#     satatment
#mastch case/ match statment
service = input("กรุณาป้อนหมายเลขบริการ:")

match service:
    case "1": print("ถอนเงิน")
    case "2": print("ฝากเงิน")
    case "3": print("สอบถามยอดเงินคงเหลือ")
    case "": print("หมายเลขบริการไม่ถูกต้อง")
```

```python
# set ไม่มีการเรียงลำดับ(ไม่มี Index)  ใช้ {} กำกับ ห้ามมีข้อมูลซ้ำ
animals = {"หมา", "แมว", "สิงโต", "ช้าง"}
print(animals) #ลำดับมันจะไม่ชัดเจน ถ้ามีข้อมูลซ้ำ มันจะเอาออกไปเลย
for item in animals:
    print(item)

#การหาข้อมูล
print("หมา" in animals) # จะแสดงผลเป็น True or False
```

```python
animals = {"หมา", "แมว", "สิงโต", "ช้าง"}
animals.add(("เป็ด"))
animals.update(("หมู","ยี่ราฟ"))
print("ข้อมูลที่ 1",animals)

pets = set(("หมา","แมว","กระต่าย"))
print("ข้อมูลที่ 2",pets)

data = animals.union(pets)
print("test1", data)
data = animals.intersection(pets)
print("test2", data)
data = animals.difference(pets)
print("test3", data)
```

```python
# Fucntion dict.
animals = {
    "หมา" : "dog",
    "แมว" : "cat",
    "สิงโต" : "Loin",
    "ช้าง" : "Elf"
}

#อยากรู้ว่าใน dict มีอะไรบ้าง
print(animals.keys())
print(animals.values())
print(animals.items()) #ได้เป็น tuple

#การเรียกข้อมูล
for key in animals.keys():
    print(key)
print("---------------------")
for value in animals.values():
    print(value)
print("---------------------")
for key,value in animals.items():
    print(value) #อยากได้อะไร เลือกอันนั้น (key:value)
print("---------------------")
for key,value in animals.items():
    print(key, "=", value)
print("---------------------")

#การดึงข้อมูลมาใช้งาน
print(animals["แมว"])
print(animals.get("แมว"))

#การจัดการข้อมูล

animals.pop("ช้าง") #ลบข้อมูลเฉพาะที่
# animals.clear() #ลบข้อมูลทั้งหมด
animals.update({"ปลา" : "Fish"}) #insert data ({key:value})
animals.update({"แมว":"kitty"}) #การแก้ไขข้อมูล
main_animal = animals.copy() #การ copy มาไว้ใน dict ใหม่
print(animals)
print(main_animal)
```

```python
'''
เป็นรูปแบบการเก็บความสัมพันธ์ของข้อมูลใน ลักษณะของ key 
และ value กล่าวคือ จะใช้ key เป็น index 
ในการเข้าถึงข้อมูล (value) แต่ละตัว
โดยค่า key นั้นต้องไม่ซ้ำกัน และสามารถกำหนดประเภทข้อมูลได้
'''
animals = {
    "หมา" : "dog",
    "แมว" : "cat",
    "สิงโต" : "Loin",
    "ช้าง" : "Elf"
}

animals["ปลา"] = "Fish" #add data
animals["ช้าง"] = "EFFT" #edit data
print(animals["หมา"])
print(animals)
```

```python
"""
operator ใช้สำหรับเปรียบเทียบข้อมูลว่าเหมือนกันหรือไม่ ?
โดยผลลัพธ์จากการเปรียบเทียบจะได้ค่าทางตรรกศาสตร์มาใช้งาน ประกอบด้วยคำสั่ง
• is เหมือนกัน
• is not ไม่เหมือนกัน
"""
#A and B as the same? identity operator
colorsA = ["red","green","white","black"]
colorsB = ["red","green","white","black"]
data = colorsA

print(colorsA is colorsB)
print(colorsA is data)
print(colorsA is not colorsB)
print(colorsA is not data)
```

```python
'''
membership_operator ใช้สำหรับเปรียบเทียบข้อมูลว่าอยู่ในกลุ่มข้อมูลหรือไม่ ?
โดยผลลัพธ์จากการเปรียบเทียบจะได้ค่าทางตรรกศาสตร์มาใช้งาน ประกอบด้วยคำสั่ง
- in อยู่ในกลุ่ม
- not in ไม่อยู่ในกลุ่ม
'''
colors = ["red","green","white","black"]
print("red" in colors)
print("red" not in colors)
```
